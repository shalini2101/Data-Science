```python
#UBCF
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("book.csv",encoding='latin1')
df
```

```python
df.head()
```

```python
book_df=df.rename({"User.ID":"userid","Book.Title":"bookttl","Book.Rating":"bookrtg"},axis='columns')
book_df
```

```python
#number of unique users in the dataset
len(book_df.userid.unique())
```

```python
len(book_df.bookttl.unique())
```

```python
book_df.duplicated(subset=None,keep='first')
```

```python
book_df.drop_duplicates(subset=None,keep='first',inplace=False,ignore_index=False)
```

```python
user_book_df = book_df.pivot_table(index='userid',
                                 columns='bookttl',
                                 values='bookrtg').reset_index(drop=True)
```

```python
user_book_df
```

```python
user_book_df.index = book_df.userid.unique()
```

```python
user_book_df
```

```python
#Impute those NaNs with 0 values
user_book_df.fillna(0, inplace=True)
```

```python
user_book_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_book_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book_df.userid.unique()
user_sim_df.columns = book_df.userid.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```

```python
#IBCF
#Fetching ratings for open water
open_water_rating = user_book_df['Twilight']  
```

```python
open_water_rating
```

```python
#Finding the correlation with different books
similar_to_open_water  = user_book_df.corrwith(open_water_rating) 
```

```python
similar_to_open_water
```

```python
corr_openwtr = pd.DataFrame(similar_to_open_water, columns=['Correlation'])
corr_openwtr.dropna(inplace=True)
corr_openwtr.head()
```

```python
corr_openwtr[corr_openwtr['Correlation'] > 0].sort_values(by='Correlation', ascending=False) 
```
