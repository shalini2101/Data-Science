```python
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
X = np.random.uniform(0,1,1000)
Y = np.random.uniform(0,1,1000)
X
```

```python
df_xy =pd.DataFrame(columns=["X","Y"])
df_xy
df_xy.X = X
df_xy.Y = Y
df_xy
df_xy.plot(x="X",y = "Y",kind="scatter")
model1 = KMeans(n_clusters=5).fit(df_xy)
```

```python
df_xy.plot(x="X",y = "Y",c=model1.labels_,kind="scatter",s=10,cmap=plt.cm.coolwarm_r)
```

```python
air = pd.read_excel("EastWestAirlines.xlsx",sheet_name='data')
air
```

```python
air.head()
```

```python
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)
```

```python
df_norm = norm_func(air.iloc[:,1:])
```

```python
df_norm.head(10)
```

```python
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(df_norm)
```

```python
y_kmeans
#clf.cluster_centers_
clf.labels_
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
air['clust']=md # creating a  new column and assigning it to new column 
air
```

```python
air.iloc[:,1:7].groupby(air.clust).mean()
```

```python
air.plot(x="Balance",y ="Qual_miles",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```
